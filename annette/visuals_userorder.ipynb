```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# reading in the data

uao_df = pd.read_csv("users_and_orders.csv")
print('size of uao_df:', uao_df.shape)
```

```python
uao_df.columns
```

```python
feat = 'date.x'
data = pd.concat([uao_df['result'], uao_df[feat]], axis = 1)
f, ax = plt.subplots(figsize = (8,8))
fig = sns.barplot(x=feat, y="result", data = data, ci = None)
fig.set(xticks = [])
```

```python
import matplotlib.ticker as ticker
feat = 'pv'
data = pd.concat([uao_df['result'], uao_df[feat]], axis = 1)
f, ax = plt.subplots(figsize = (16,12))
#ax.set_xticks([])
fig = sns.barplot(x=feat, y="result", data = data, )
fig.axis(ymin = 0, ymax = .5);
fig.axis(xmin = 0, xmax = 200);
plt.xlabel("Page Views")
plt.ylabel("Average Approval Rate")
fig.set(xticks = [])
```
