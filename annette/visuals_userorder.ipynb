```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# reading in the data

uao_df = pd.read_csv("users_and_orders.csv")
print('size of uao_df:', uao_df.shape)
```

```python
feat = 'date.x'
data = pd.concat([uao_df['result'], uao_df[feat]], axis = 1)
f, ax = plt.subplots(figsize = (8,8))
fig = sns.barplot(x=feat, y="result", data = data)
```
